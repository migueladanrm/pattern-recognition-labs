Preparaciones preliminares del entorno

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import PIL

def open_image(path):
    img = cv.imread(path)
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)

def display_image(img, title):
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")
```

## 3.1 Balance de color

Escriba un programa que permita cambiar el balance de color de la imagen multiplicando cada valor de color por un n√∫mero constante del usuario

```python
# Carga de la imagen
img = open_image("data/naranjas1.tif")

# RGB de la imagen
R, G, B = cv.split(img)

# Par√°metro del usuario (debe ser un valor entre 0 y 255)
c = 3

# Se fusionan los canales RGB multiplicados por la constante del usuario.
result = cv.merge((R * c, G * c, B * c))

display_image(result, "naranjas")
```

## 3.3 Remover fondo
En este ejercicio se remueve el fondo verde de una imagen y se deja √∫nicamente el objeto sin fondo.

```python
# se carga el fondo y la imagen
background = open_image("data/fondo-verde.png")
img = open_image("data/naranja-fondo-verde.png")

# se calcula la diferencia de las im√°genes
diff = cv.bitwise_xor(img, background)

display_image(diff,"recorte")
```

## 3.5 Efectos de fotograf√≠a
En este ejercicio se aplican filtros de contraste.

```python
def contrast(img, alpha, beta):
    R, G, B = cv.split(img)
    return cv.merge((R * alpha + beta, G * alpha + beta, B * alpha + beta))

img = open_image("data/naranjas1.tif")

display_image(contrast(img, 1, 128), "resultado")
```

## 3.6 Equalizaci√≥n de histograma

```python
# carga de la imagen
img = open_image("data/naranjas1.tif")

#canales de color originales
R, G, B = cv.split(img)

# canales de color equalizados
R_e = cv.equalizeHist(R)
G_e = cv.equalizeHist(G)
B_e = cv.equalizeHist(B)

equalized = cv.merge((R_e, G_e, B_e))

display_image(np.hstack((img, equalized)), "original/equalizado")
```
